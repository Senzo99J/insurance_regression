```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split

# Import the dataset
df = pd.read_csv("insurance (1).csv")
df.head()
```

```python
# Extract relevant columns

ages = df.iloc[:,:1].values
charges = df.iloc[:,-1].values

# Plot scatter plot g
plt.figure(figsize=(10, 6))
plt.scatter(ages, charges, color='blue', alpha=0.5)
plt.title('Age vs Insurance Charges')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.show()
```

```python
# Fit linear regression model
model = LinearRegression()
model.fit(ages, charges)

# Make predictions
charges_pred = model.predict(ages)

# Plot scatter plot with best-fit line
plt.figure(figsize=(10, 6))
plt.scatter(ages, charges, color='blue', alpha=0.5, label='Data points')
plt.plot(ages, charges_pred, color='red', label='Best-fit line')
plt.title('Age vs Insurance Charges with Best-Fit Line')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.legend()
plt.show()
```
